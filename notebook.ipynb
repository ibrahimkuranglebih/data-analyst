### Pertanyaan 2:

```python
polutan_kota_Aotizhongxin = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
grouped_data = df_aothi.groupby('year')[polutan_kota_Aotizhongxin].mean()
plt.figure(figsize=(12, 6))
for polutan in polutan_kota_Aotizhongxin:
    plt.plot(grouped_data.index, grouped_data[polutan], marker='o', label=polutan)
plt.title('Rata-Rata Tiap Polutan per Tahun', fontsize=16)
plt.xlabel('Tahun', fontsize=14)
plt.ylabel('Konsentrasi Rata-Rata', fontsize=14)
plt.legend(title="Polutan")
plt.grid(True)
plt.tight_layout()
plt.show()
```
